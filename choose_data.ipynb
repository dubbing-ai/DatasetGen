```python
#import csv to pandas
import pandas as pd
# Create a DataFrame from a CSV file
df = pd.read_csv('generated/processed/combined_sentences_with_phoneme.csv')
df
```

```python
sentences = df['thai phoneme'].tolist()
splitsentences=[]
for sentence in sentences:
    splitsentence=sentence.split(' ')
    for i in splitsentence:
        if i == '_':
            splitsentence.remove(i)
    unionlist=set(splitsentence)
    splitsentence=list(unionlist)
    splitsentences.append(splitsentence)
splitsentenceswithindex=[]
for i,sentence in enumerate(splitsentences):
    splitsentenceswithindex.append((i,sentence))
all_phonemes=set()
for line in splitsentences:
    for phoneme in line:
        all_phonemes.add(phoneme)
print('phoneme num:',len(all_phonemes),all_phonemes)
```

```python
def select_lines(lines,all_phonemes):
    covered_phonemes = set()
    selected_indices = []

    while covered_phonemes != all_phonemes:
        best_index = None
        max_uncovered = 0
        min_total_phonemes = float('inf')

        for i, line in enumerate(lines):
            index, value = line
            
            if index in selected_indices:
                continue

            line_set = set(value)
            uncovered = line_set - covered_phonemes
            num_uncovered = len(uncovered)
            num_total = len(value)

            # Select the line maximizing uncovered phonemes and minimizing total phonemes
            if num_uncovered > max_uncovered or (num_uncovered == max_uncovered and num_total < min_total_phonemes):
                best_index = index
                best_value = value
                max_uncovered = num_uncovered
                min_total_phonemes = num_total

        if best_index is not None:
            selected_indices.append(best_index)
            for phoneme in best_value:
                covered_phonemes.add(phoneme)
        
        else:
            print("Breaking - There is no best index")
            return selected_indices

    return selected_indices

result = select_lines(splitsentenceswithindex, all_phonemes)
print("Selected line indices:", result)
```

# cut result before greedy again

```python
result_set = set()
result_set.add(tuple(result))
breakcount = 0
num_dataset = 5 # number of dataset
while len(result_set)!=num_dataset:
    for index in result:
        for line in splitsentenceswithindex:
            if line[0]==index:
                splitsentenceswithindex.remove(line)
    print(len(splitsentenceswithindex))
    result = select_lines(splitsentenceswithindex, all_phonemes)
    print("Selected line indices:", result)
    #check
    check_set=set()
    count=0
    for index in result:
        for phoneme in splitsentences[index]:
            count+=1
            check_set.add(phoneme)
    print('check phoneme:',len(check_set),len(all_phonemes),check_set==all_phonemes)
    print('num phoneme:',count)
    print('num line:',len(result))
    print('----------------------------------------------------------------')
    if check_set==all_phonemes:
        result_set.add(tuple(result))
    breakcount += 1
    if breakcount==num_dataset:
        break

print('num_dataset:',len(result_set),result_set)
```

# export

```python
# export to folder
import os

# Create the folder if it doesn't exist
output_folder = "greedy_data"
os.makedirs(output_folder, exist_ok=True)

c = 1
for i in result_set:
    result = list(i)
    print(result)
    df_result = df.iloc[result]
    df_result = df_result[['thai sentence', 'thai phoneme']]
    csv_filename = os.path.join(output_folder, f'greedy_{c}.csv')
    df_result.to_csv(csv_filename, index=False)
    c += 1

print(f"All CSV files are saved in the '{output_folder}' folder.")
```
